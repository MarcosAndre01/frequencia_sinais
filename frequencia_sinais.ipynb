```python
import pandas as pd
from collections import Counter
```

```python
corpus = pd.read_csv("corpus.csv", header=None)
corpus.rename(columns={0: "Portugues", 1: "Glossa"}, inplace=True)
corpus
```

```python
sinais_vlibras = pd.read_json("sinais.json")
sinais_vlibras.rename(columns={0: "Sinal"}, inplace=True)
sinais_vlibras = sinais_vlibras.assign(
    TemNoVLibras=True,
)
sinais_vlibras
```

```python
freq = Counter()
corpus["Glossa"].str.split().apply(freq.update)
# sinais_glossa = freq.keys()
# sinais_glossa

# freq = pd.DataFrame([freq])
# freq

freq = pd.DataFrame.from_dict(freq, orient="index").reset_index()
freq.rename(columns={"index":"Sinal", 0:"Frequencia"}, inplace=True)
freq
```

```python
freq = freq.merge(sinais_vlibras, on='Sinal', how='outer')
freq['TemNoVLibras'] = freq['TemNoVLibras'].fillna(False)
freq['Frequencia'] = freq['Frequencia'].fillna(0)
```
